# training models

```python
from gwp.train import load_data,training_models,features_analysis
```

```python
X_train, y_train, X_test, y_test = load_data("featuresforgwp.csv", "divided_set_0.8_0.2_data_split.txt")
```

```python
training_models(X_train, y_train.T[0], X_test, y_test.T[0], model_name = "LF")
```

```python
training_models(X_train, y_train.T[1], X_test, y_test.T[1], model_name = "GWP")
```

```python
features_analysis("LF.pkl", "featuresforgwp.csv")
```

```python
features_analysis("GWP.pkl", "featuresforgwp.csv")
```
